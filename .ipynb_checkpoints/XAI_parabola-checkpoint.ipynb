```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

# Funzione per generare dati in base all'equazione del moto uniformemente accelerato
def generate_data(s_0, v_0, a, t_values):
    s_values = s_0 + v_0 * t_values + 0.5 * a * t_values**2
    return s_values

# Parametri iniziali
s_0 = 10.0  # Posizione iniziale
v_0 = 2.0   # Velocit iniziale
a = 0.5     # Accelerazione

# Genera dati di addestramento
np.random.seed(42)
num_samples = 100
t_train = np.sort(np.random.uniform(0, 5, num_samples))
s_train = generate_data(s_0, v_0, a, t_train)

# Crea il modello della rete neurale
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(1,), activation='relu'),
    keras.layers.Dense(1)
])

# Compila il modello
model.compile(optimizer='adam', loss='mean_squared_error')

# Addestra il modello
history = model.fit(t_train, s_train, epochs=300, verbose=0)

# Genera dati di test
t_test = np.array([3.0])  # Punto di test

# Esegui predizione
s_test = model.predict(t_test)

# Visualizza i risultati
plt.figure(figsize=(12, 6))

# Grafico dei dati di addestramento
plt.subplot(1, 2, 1)
plt.scatter(t_train, s_train, label='Dati di addestramento')
plt.scatter(t_test, s_test, color='red', label='Dato di test')  # Punto di test in rosso
plt.xlabel('Tempo (t)')
plt.ylabel('Posizione (s)')
plt.title('Dati di Addestramento')
plt.legend()

# Grafico della loss durante l'addestramento
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.xlabel('Epoca')
plt.ylabel('Loss')
plt.title('Andamento della Loss durante l\'Addestramento')

plt.tight_layout()
plt.show()

# Stampa il risultato della predizione
print(f'Punto di test: t={t_test[0]}, Posizione stimata={s_test[0, 0]}')
```
